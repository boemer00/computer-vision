```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = os.path.join('raw_data')
train_ship_path = os.path.join(file_path, 'train', 'ship')
train_no_ship_path = os.path.join(file_path, 'train', 'no-ship')
```

```python
train_ship_images = os.listdir(train_ship_path)
train_no_ship_images = os.listdir(train_no_ship_path)
```

```python
train_ship_images[:3]
```

```python
for img in range(3):
    img_path = os.path.join(train_ship_path, train_ship_images[img])
    img = cv2.imread(img_path)

    if img is not None:
        print(img.shape)
    else:
        print(f"Image at {img_path} could not be loaded.")

    plt.imshow(img)
    plt.show()
```
